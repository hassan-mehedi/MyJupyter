# Importing Libraries
* To access classifiers
* To fit training data in classifiers
* To test the model using test data
* To evaluate the models

```python
#Importing Regressors
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

#Importing metrics to calculate error
from sklearn.metrics import confusion_matrix

#Importing to train and test data
from sklearn.model_selection import train_test_split
import time
```

# Separating target field from other fields

```python
X = df.drop("label", axis = 1)
y = df["label"]
```

# Training and Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Model Evaluation
* Initialize Model
* Fitting data to the model
* Testing the model
* Find out the accuracy of the model

## RandomForestClassifier

```python
from sklearn.metrics import accuracy_score
rfc_score_list = []
rfc_exec_time_list = []
est = 100
for i in range(100):
    rfc = RandomForestClassifier(n_estimators=est)
    rfc.fit(X_train,y_train)
    start_time = time.time()
    rfc_pred = rfc.predict(X_test)
    end_time = time.time()
    rfc_exec_time_list.append(end_time - start_time)
    rfc_score_list.append(accuracy_score(y_test, rfc_pred))
    est = est + 10
rfc_exec_time_list = np.array(rfc_exec_time_list)
rfc_score_list = np.array(rfc_score_list)
accuracy_score = np.mean(rfc_score_list)
print("Accuracy score of Random Foret Classifier: ", accuracy_score)
```

```python
accucary_score_df = pd.DataFrame(rfc_score_list, columns=["RandomForestClassifier"])
exec_time_df = pd.DataFrame(rfc_exec_time_list, columns=["RandomForestClassifier"])
```

```python
exec_time_df[:5]
```

```python
accucary_score_df[:5]
```

## AdaBoostClassifier

```python
from sklearn.metrics import accuracy_score
abc_score_list = []
abc_exec_time_list = []
est = 100
for i in range(100):
    abc = AdaBoostClassifier(n_estimators=est)
    abc.fit(X_train,y_train)
    start_time = time.time()
    abc_pred = abc.predict(X_test)
    end_time = time.time()
    abc_exec_time_list.append(end_time - start_time)
    abc_score_list.append(accuracy_score(y_test, abc_pred))
    est = est + 10
abc_exec_time_list = np.array(abc_exec_time_list)
abc_score_list = np.array(abc_score_list)
accuracy_score = np.mean(abc_score_list)
print("Accuracy score of AdaBoost Classifier: ", accuracy_score)
```

```python
accucary_score_df["AdaBoost Classifier"] = abc_score_list
exec_time_df["AdaBoost Classifier"] = abc_exec_time_list
```

```python
exec_time_df[:5]
```

## DecisionTreeClassifier

```python
from sklearn.metrics import accuracy_score
dtc_score_list = []
dtc_exec_time_list = []
for i in range(100):
    dtc = DecisionTreeClassifier(random_state=i)
    dtc.fit(X_train,y_train)
    start_time = time.time()
    dtc_pred = dtc.predict(X_test)
    end_time = time.time()
    dtc_exec_time_list.append(end_time - start_time)
    dtc_score_list.append(accuracy_score(y_test, dtc_pred))
dtc_exec_time_list = np.array(dtc_exec_time_list)
dtc_score_list = np.array(dtc_score_list)
accuracy_score = np.mean(dtc_score_list)
print("Accuracy score of Decision Tree Classifier: ", accuracy_score)
```

```python
accucary_score_df["Decision Tree Classifier"] = dtc_score_list
exec_time_df["Decision Tree Classifier"] = dtc_exec_time_list
```

## MLPClassifier

```python
from sklearn.metrics import accuracy_score
mlp_score_list = []
mlp_exec_time_list = []
for i in range(100):
    mlp = MLPClassifier(random_state=i, max_iter=300)
    mlp.fit(X_train,y_train)
    start_time = time.time()
    mlp_pred = mlp.predict(X_test)
    end_time = time.time()
    mlp_exec_time_list.append(end_time - start_time)
    mlp_score_list.append(accuracy_score(y_test, mlp_pred))
mlp_exec_time_list = np.array(mlp_exec_time_list)
mlp_score_list = np.array(mlp_score_list)
accuracy_score = np.mean(mlp_score_list)
print("Accuracy score of MLP Classifier: ", accuracy_score)
```

## KNeighborsClassifier

```python
from sklearn.metrics import accuracy_score
knc_score_list = []
knc_exec_time_list = []
neigh = 5
for i in range(100):
    knc = KNeighborsClassifier(n_neighbors=neigh)
    knc.fit(X_train,y_train)
    start_time = time.time()
    knc_pred = knc.predict(X_test)
    end_time = time.time()
    knc_exec_time_list.append(end_time - start_time)
    knc_score_list.append(accuracy_score(y_test, knc_pred))
    neigh = neigh + 1
    if neigh > 15:
        neigh = 5
knc_exec_time_list = np.array(knc_exec_time_list)
knc_score_list = np.array(knc_score_list)
accuracy_score = np.mean(knc_score_list)
print("Accuracy score of knc Classifier: ", accuracy_score)
```

# Defining function to visualize Confusion Matrix for Each Model Prediction

```python
def plot_confusion_metrix(res, name):
    df_cm = pd.DataFrame(res, columns=np.unique(y_test), index = np.unique(y_test))
    df_cm.index.name = 'Actual'
    df_cm.columns.name = 'Predicted by {}'.format(name)
    plt.figure(figsize = (15,10))
    sns.set(font_scale=1.6)
    sns.heatmap(df_cm, cmap="Blues", annot=True,annot_kws={"size": 16});
    
```
